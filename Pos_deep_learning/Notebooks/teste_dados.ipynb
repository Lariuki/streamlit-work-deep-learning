```python
from pathlib import Path
import pandas as pd

nome_arquivo = "base_sintetica_tcc_10234_anonimizada.csv"
base_busca = Path.cwd()

candidatos = [
    base_busca / "Dados" / nome_arquivo,
    base_busca / nome_arquivo,
]

for parent in [base_busca, *base_busca.parents]:
    caminhos_possiveis = [
        parent / "Dados" / nome_arquivo,
        parent / nome_arquivo,
    ]
    candidatos.extend(caminhos_possiveis)

encontrados = [c.resolve() for c in candidatos if c.exists()]
csv_path = encontrados[0] if encontrados else None

print(f"Diretório atual do kernel: {base_busca}")

if csv_path is None:
    raise FileNotFoundError(
        "Arquivo não encontrado neste ambiente. "
        f"O kernel está rodando em: {base_busca}. "
        "Se você estiver no Colab ou em outro ambiente remoto, esse caminho do seu Mac não existe lá. "
        "Nesse caso, faça upload do CSV para o ambiente atual ou monte a pasta do projeto."
    )

# Detectar separador automaticamente
separadores_candidatos = [';', ',', '\t', '|']
df = None

for sep in separadores_candidatos:
    try:
        df_test = pd.read_csv(csv_path, sep=sep, quotechar='"', nrows=5)
        # Verificar se o número de colunas faz sentido (mais 1 do que sem separador)
        if len(df_test.columns) > 1:
            print(f"✅ Separador detectado: '{sep}'")
            df = pd.read_csv(csv_path, sep=sep, quotechar='"')
            break
    except:
        continue

if df is None:
    raise ValueError(f"Não foi possível parsejar o CSV com nenhum separador comum: {separadores_candidatos}")

print(f"\nArquivo carregado: {csv_path}")
print(f"Linhas: {df.shape[0]} | Colunas: {df.shape[1]}")
print("\nColunas:")
print(df.columns.tolist())

df.head()
```
